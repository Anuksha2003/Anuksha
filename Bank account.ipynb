```python
class Bank_Account():
    def __init__(self,name,number,typeac,amount):
        self.name=name
        self.number=number
        self.type=typeac
        self.amount=amount
        
    def display(self):
        print("The account holder is:{}".format(self.name))
        print("The account number is:{}".format(self.number))
        print("The type of account is:{}".format(self.type))
        print("The amount is:{}".format(self.amount))
        
    def deposit(self,depo):
        self.amount=self.amount+depo

    
    def withdraw(self,withd):
        self.amount=self.amount-withd
    
    def check_balance(self):
        return self.amount
    
print("Enter the account details:")
Name=input("Enter the name of Account Holder:")

Accno=int(input("Enter the account number:"))
print(""" Press
1-Savings account
2- Salary account
3-Current account
4-Recurring deposit account
5-NRI Account""")
n=int(input("Enter the option:"))
if n==1:
    Acctype="Savings account"
elif n==2:
    Acctype="Salary account"
elif n==3:
    Acctype="Current account"
elif n==4:
    Acctype="Recurring deposit account"
elif n==5:
    Acctype="NRI Account"
else:
    print("Invalid Input")
    
Balance=float(input("Enter the amount to be deposited in the account:"))

BankAcc=Bank_Account(Name,Accno,Acctype,Balance)
print(""""Press
1-for depositing amount
2-withdrawing amount
3-checking the balance
4-display details
0- quit""")
while 1:
    p=int(input("Enter the option:"))
    if p==1:
        dep=int(input("Enter amount to be deposited:"))
        BankAcc.deposit(dep)
    elif p==2:
        wth=int(input("Enter amount to be withdrawn:"))
        if(BankAcc.check_balance()>wth):
            BankAcc.withdraw(wth)
        else :
            print("Insufficient balance")
    elif p==3:
        print("The balance in your account is {}".format(BankAcc.check_balance()))
    elif p==4:
        BankAcc.display()
    elif p==0:
        break
    else:
        print("Error")
print("Thankyou")

    
```
