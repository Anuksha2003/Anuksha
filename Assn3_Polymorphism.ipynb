```python
#Polymorphism in class methods
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print(f"I am a cat. My name is {self.name}. I am {self.age} years old.")

    def sound(self):
        print("Meow")


class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print(f"I am a dog. My name is {self.name}. I am {self.age} years old.")

    def sound(self):
        print("Bark")


cat1 = Cat("Kitty", 5)
dog1 = Dog("Fluffy", 4)

for animal in (cat1, dog1):
    animal.info()
    animal.sound()
```

```python
#Method Overriding
class polygon():
    def area(self):
        pass
    def perimeter(self):
        pass
class rectangle(polygon):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        return (self.length*self.breadth)
    def perimeter(self):
        return (2*(self.length+self.breadth))
class square(polygon):
    def __init__(self,side):
        self.side=side
    def perimeter(self):
        return (4*self.side)
    def area(self):
        return (self.side*self.side)
class triangle(polygon):
    def __init__(self,s1,s2,s3):
        self.s1=s1
        self.s2=s2
        self.s3=s3
    def area(self):
        s=(s1+s2+s3)/2;
        return ((s*(s-s1)*(s-s2)*(s-s3))**0.5)
    def perimeter(self):
        return (s1+s2+s3)

print("Enter 1 for rectangle, 2 for square and 3 for triangle:")
choice=int(input())
if (choice==1):
    print("Enter length and breadth:")
    l=eval(input())
    b=eval(input())
    rect1=rectangle(l,b)
    print("Enter 1 for area and 2 for perimeter:")
    ch1=int(input())
    if (ch1==1):
        print(rect1.area())
    elif (ch1==2):
        print(rect1.perimeter())
    else:
        print("Invalid Input")
elif (choice==2):
    print("Enter side:")
    s=eval(input())
    sq1=square(s)
    print("Enter 1 for area and 2 for perimeter:")
    ch2=int(input())
    if (ch2==1):
        print(sq1.area())
    elif (ch2==2):
        print(sq1.perimeter())
    else:
        print("Invalid Input")
elif (choice==3):
    print("Enter base and height:")
    b=eval(input())
    h=eval(input())
    tr1=triangle(b,h)
    print("Enter 1 for area and 2 for perimeter:")
    ch3=int(input())
    if (ch3==1):
        print(tr1.area())
    elif (ch3==2):
        print(tr1.perimeter())
    else:
        print("Invalid Input")
else:
    print("Invalid Input")

    
```
